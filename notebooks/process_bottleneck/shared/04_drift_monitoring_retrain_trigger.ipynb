```python
#cell 2
def psi(expected: np.ndarray, actual: np.ndarray, bins: int = 10) -> float:
    expected = expected[np.isfinite(expected)]
    actual = actual[np.isfinite(actual)]
    if len(expected) == 0 or len(actual) == 0:
        return 0.0
    q = np.quantile(expected, np.linspace(0,1,bins+1))
    q[0] -= 1e-9
    q[-1] += 1e-9

    e_counts, _ = np.histogram(expected, bins=q)
    a_counts, _ = np.histogram(actual, bins=q)

    e_perc = (e_counts / max(1, e_counts.sum())) + 1e-6
    a_perc = (a_counts / max(1, a_counts.sum())) + 1e-6
    return float(np.sum((a_perc - e_perc) * np.log(a_perc / e_perc)))

# Ví dụ drift duration của 1 step
step_col = "STEP_12_TRANSIT_duration_min"  # đổi theo step code của registry bạn
# (nếu step code khác, bạn mở registry để lấy đúng tên)

# Bạn sẽ dùng features đã build (hoặc load từ features csv nếu bạn lưu)
```
